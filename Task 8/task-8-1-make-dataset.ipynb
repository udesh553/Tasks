```python
 
players_df=pd.read_csv("../input/nfl-big-data-bowl-2023/players.csv")
pffScoute_df=pd.read_csv("../input/nfl-big-data-bowl-2023/pffScoutingData.csv")
games_df=pd.read_csv("../input/nfl-big-data-bowl-2023/games.csv")
plays_df=pd.read_csv("../input/nfl-big-data-bowl-2023/plays.csv")

```

```python
# Tracking data: Files week[week].csv contain player tracking data from season [week].
# The key variables are gameId, playId, and nflId.
W1_df=pd.read_csv("../input/nfl-big-data-bowl-2023/week1.csv")
W2_df=pd.read_csv("../input/nfl-big-data-bowl-2023/week2.csv")
W3_df=pd.read_csv("../input/nfl-big-data-bowl-2023/week3.csv")
W4_df=pd.read_csv("../input/nfl-big-data-bowl-2023/week4.csv")
W5_df=pd.read_csv("../input/nfl-big-data-bowl-2023/week5.csv")
W6_df=pd.read_csv("../input/nfl-big-data-bowl-2023/week6.csv")
W7_df=pd.read_csv("../input/nfl-big-data-bowl-2023/week7.csv")
W8_df=pd.read_csv("../input/nfl-big-data-bowl-2023/week8.csv")
```
